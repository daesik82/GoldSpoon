```python
import BigBull
import pandas as pd
```

```python
conn_string = "사용자 입력"
```

```python
db = BigBull.StockDB(conn_string, BigBull.__meta)
```

### 재무제표 크롤러 객체 생성 

```python
crawler = BigBull.CompanyInformationCrawler(db)
```

### 전체 회사 재무제표 가져와서 저장하기

```python
fail, company_information_data = crawler.save_company_information() 
# 인자로 year=[] 원하는 년도를 리스트 요소로 넘겨주면 주면 됨. 모든 년도의 데이터가 필요하면 인자없이 함수 실행.
```

# 저장된 데이터 불러오는 여러가지 방법 

## 1. 재무제표 테이블(company_info)에 저장된 데이터를 query로 가져오기 

```python
rp = db.select_columns('company_info')
for row in rp:
    print(row.code, row.year) # 원하는 컬럼값 입력해서 불러오기
```
