```python
import BigBull
```

```python
# DB connection setting
conn_string = 'postgresql+psycopg2://daesikkim@localhost:5432/db_goldspoon'
db = BigBull.StockDB(conn_string, BigBull.__meta)
```

```python
# Crawler Settings 
stock_crawler = BigBull.StockCodeCrawler(db)
hist_crawler = BigBull.HistoryCrawler(db)
finance_crawler = BigBull.CompanyInformationCrawler(db)
```

```python
# Crawl Stocks info 
stock_crawler.save_stock_code()
```

```python
# Crawl Historical data 
stock_ls = [i for i in db.load_all_stock_code()]
names, codes = zip(*stock_ls)
```

```python
count = 0
for name in names:
    hist_crawler.save_history_of(name, hard_mode=True)
    count += 1 
    print(f"---- {name} crawled / {count/len(names)*100}% ----")
```

```python
# update DB for historical data 
count = 0 
for name in names:
    hist_crawler.save_history_of(name, start_date="2018-4-16")
    count +=1 
    print(f'---- {name} crawled / {count} / {len(names)}----')
```

```python
# Crawl Company Financial Data
finance_crawler.save_company_information()
```
