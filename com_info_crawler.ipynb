```python
conn_string = "postgresql+psycopg2://daesikkim@localhost:5432/db_goldspoon"
db = BigBull.StockDB(conn_string, BigBull.__meta)
```

```python
stocks = [i for i in db.load_all_stock_code()]
```

```python
names, codes = zip(*stocks)
```

```python
soup.select('#comInfo > tr:nth-of-type(1) > td > span:nth-of-type(6)')[0].text.split(' : ')
```

```python
company_info_dict = {'code':[], 'name':[], 'market': [], 'ics': [], 'hq_address':[], 'homepage':[], 'employee':[]}
```

```python
error_codes = []
for name, code in stocks:
    url = f'http://wisefn.stock.daum.net/company/c1020001.aspx?cmp_cd={code}&frq=&rpt='
    req = requests.get(url)
    html = req.text 
    soup = BeautifulSoup(html, 'lxml')
    
    try: 
        market, ics = soup.select('#comInfo > tr:nth-of-type(1) > td > span:nth-of-type(6)')[0].text.split(' : ')
        hq_address = soup.select('#cTB201 > tr:nth-of-type(1) > td')[0].text
        homepage = soup.select('#cTB201 > tr:nth-of-type(2) > td')[0].text
        employee = soup.select('#cTB201 > tr:nth-of-type(4) > td.c4')[0].text
    
        company_info_dict['code'].append(code)
        company_info_dict['name'].append(name)
        company_info_dict['market'].append(market)
        company_info_dict['ics'].append(ics)
        company_info_dict['hq_address'].append(hq_address)
        company_info_dict['homepage'].append(homepage)
        company_info_dict['employee'].append(employee)
        
        print(f"""
              ------------------------
              code : {code}, 
              name : {name},
              market : {market},
              ics : {ics},
              hq_address : {hq_address}, 
              homepage : {homepage}, 
              employee : {employee}
              ------------------------
              """)
        
    except: 
        error_codes.append(name)
```

```python
df_com_info = pd.DataFrame.from_dict(company_info_dict)
```
